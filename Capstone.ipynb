```python
Gender_dummies  = pd.get_dummies(new['Gender '], prefix = 'Gender ' )
new = pd.concat([new, Gender_dummies], axis = 1)
print(new)
```

```python
new['Grade'] = new['Grade'].astype('category')
new['Extracurricular Activities'] = new['Extracurricular Activities'].astype('category')
new['Nationality'] = new['Nationality'].astype('category')	
print(new.dtypes)
```

```python
Grade_dummies  = pd.get_dummies(new['Grade'], prefix = 'Grade')
new = pd.concat([new, Grade_dummies], axis = 1)
print(new)
```
